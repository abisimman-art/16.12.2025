### elif:


```python
x=int(input("enter a number:"))
if x==2:
    print("1st if")
elif x<=4:
    print("2nd if")
elif x<6:
    print("3rd if")
else:
    print("exit")
```

### nested if

```python
x=5
if x>0:
    print("1s if")
    if x>=2:
        print("2nd if")
        if x==5:
            print("3rd if")
else:
    print("else if")
```

### date time library


```python
from datetime import datetime as date
today = date.now()
print(today)
```

```python
tdy=today.strftime("%T")
print(tdy)
```

```python
tdy_date=today.strftime("%d/%m/%Y")
print(tdy_date)
```

```python
from datetime import timedelta as td
past=today-td(6000)
print(past)
```

```python
from datetime import timedelta as td
future=today+td(10000)
print(future)
```
